# (a) i.

In PS0 its proved

$$x^TAx = \sum_{i,j}^{m} A_{ij}x_i x_j$$

So,

\begin{align*}
J(\theta) &= (X\theta - \vec{y})^T W (X\theta - \vec{y}) \\
                &= \sum_{ij}^{m} W_{ij} (X\theta - \vec{y})_{i} (X\theta - \vec{y})_{j} \\
                &= \sum_{ij}^{m} W_{ij} (\theta^T x^{(i)} - y^{(i)}) (\theta^T x^{(j)} - y^{(j)}) \\
\end{align*}

where $x^{(i)}$ is the $i^{th}$ column of the matrix $X^T$, and $y^{(i)}$ is the $i^{th}$ element of $\vec{y}$.

Given $W$ is diagonal matrix of shape $m \times m$ (correspond to $m$ data points), all elements of $J(\theta)$ becomes $0$ unless when $i=j$

\begin{align*}
J(\theta) &= \sum_{i}^{m} W_{ii} (\theta^T x^{(i)} - y^{(i)})^2 \\ 
\end{align*}


To make $J(\theta)$ match 

\begin{align*}
J(\theta) &= \frac{1}{2}\sum_{i}^{m} w_{i} (\theta^T x^{(i)} - y^{(i)})^2 \\ 
\end{align*}

set $$W_{ii} = \frac{1}{2}w_{i}$$

# (a) ii.

Do the normal case for exercise, using $\nabla_x x^T A x = 2Ax$,

\begin{align*}
\nabla_{\theta} J(\theta)
    &= 2 (W (X\theta - \vec{y}))^T X \\
\end{align*}

Let $\nabla_{\theta} J(\theta) = 0$,

\begin{align*}
   2 (W (X\theta - \vec{y}))^T X &= 0 \\
   X^T(W (X\theta - \vec{y})) &= 0 \\
   X^T W X\theta - X^T W \vec{y} &= 0 \\
   X^T W X\theta &= X^T W \vec{y} \\
   \theta &= (X^T W X)^{-1} X^T W \vec{y} \\
\end{align*}

When $W = I$, i.e. all data points are equally weighted,

$$\theta = (X^T X)^{-1} X^T \vec{y}$$

which is exactly the normal case.

# (a) iii.

\begin{align*}
l &= \mathrm{log} \prod_{i=1}^{m} p(y^{(i)}|x^{(i)}; \theta) \\
  &= \mathrm{log} \prod_{i=1}^{m} \frac{1}{\sqrt{2\pi} \sigma^{(i)}} \mathrm{exp}\bigg (- \frac{(y^{(i)} - \theta^Tx^{(i)})^2}{2(\sigma^{(i)})^2} \bigg) \\
  &= \sum_{i=1}^{m} \mathrm{log}\frac{1}{\sqrt{2\pi} \sigma^{(i)}} - \frac{1}{2} \sum_{i=1}^{m} \frac{1}{(\sigma^{(i)})^2} (y^{(i)} - \theta^Tx^{(i)})^2
\end{align*}

Comparing $l$ to $J(\theta) = \frac{1}{2}\sum_{i}^{m} w_{i} (\theta^T x^{(i)} - y^{(i)})^2$,

1. the first part of $l$ is a constant, so ignore it. 
1. maximizing $l$ is equivalent to minimizing the negative of it: $\frac{1}{2} \sum_{i=1}^{m}\frac{1}{(\sigma^{(i)})^2} (y^{(i)} - \theta^Tx^{(i)})^2$
1. so $w_i$ can be written as $\frac{1}{(\sigma^{(i)})^2}$

Therefore, maxmizing this likelihood is equivalent to solving a weighted linear regression problem.

Implement $\theta = (X^T X)^{-1} X^T \vec{y}$ for normal cases, and $\theta = (X^T W X)^{-1} X^T W \vec{y}$ for general cases.

```python
def optim(X, y, W=None):
    if W is None:
        return np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
    else:
        return np.linalg.inv(X.T.dot(W).dot(X)).dot(X.T).dot(W).dot(y)
```

```python
x0 = np.ones(df_tv.shape[1])
x1 = wave_lens
y = df_tv.head(1).values.ravel()
X = np.stack([x0, x1]).T
```

```python
theta = optim(X, y)
```

# (b) ii.

```python
def construct_weight_matrix(X, x_eval, tau=5):
    return np.diag(np.exp(- (X[:,1] - x_eval[1]) ** 2 / (2 * tau ** 2)))
```

```python
x0 = np.ones(df_tv.shape[1])
x1 = wave_lens
y = df_tv.head(1).values.ravel()
X = np.stack([x0, x1]).T
```

```python
res = []
preds = []
for k, x_eval in enumerate(X):
    W = construct_weight_matrix(X, x_eval)
    theta = optim(X, y, W)
    preds.append(theta.dot(x_eval))
```

```python
plt.plot(wave_lens, df_tv.loc[0].values)
plt.plot(wave_lens, preds)
plt.show()
```

# (b) iii.

Varying tau values and see its effects

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for k, tau in enumerate([1, 10, 100, 1000]):
    ax = axes[k]
    preds = []
    for x_eval in X:
        W = construct_weight_matrix(X, x_eval, tau=tau)
        theta = optim(X, y, W)
        preds.append(theta.dot(x_eval))
    
    ax.plot(wave_lens, df_tv.loc[0].values)
    ax.plot(wave_lens, preds)
    ax.set_title('tau = {0}'.format(tau))

plt.show()
```

So when $\tau$ is small, the prediction becomes very sensitive to the distance between the data point to be evaluated and the training data. Hence, the prediction also appears to be noisy. When $\tau$ is big, the opposite happens.

# (c) i.

Smooth all spectra

```python
def lwr_single(X, y, x, tau):
    '''Locally weighted regression for a single observation'''
    W = construct_weight_matrix(X, x, tau=tau)
    theta = optim(X, y, W)
    return theta
```

```python
def lwr(X, y, xs, tau):
    '''lwr for multiple xs'''
    thetas = []
    for x in xs:
        th = lwr_single(X, y, x, tau)
        thetas.append(th)
    return thetas
```

```python
# Smoothed Spectrum
fs_tv = []
x0 = np.ones(df_tv.shape[1])
x1 = wave_lens
X = np.stack([x0, x1]).T
# used for calculating distances d(f1, f2)
for k, row in df_tv.iterrows():
    print(k, end=',')
    y = row.values
    thetas = lwr(X, y, X, tau=5)
    prd = [_t.dot(_x) for (_t, _x) in zip(thetas, X)]
    fs_tv.append(prd)

df_fs_tv = pd.DataFrame(fs_tv, columns=df_tv.columns)
```

```python
df_fs_tv.shape
```

```python
# smoothed spectrum (Test)
fs_test = []
x0 = np.ones(df_test.shape[1])
x1 = wave_lens
X = np.stack([x0, x1]).T
for k, row in df_test.iterrows():
    print(k, end=',')
    y = row.values
    thetas = lwr(X, y, X, tau=5)
    prd = [_t.dot(_x) for (_t, _x) in zip(thetas, X)]
    fs_test.append(prd)
df_fs_test = pd.DataFrame(fs_test, columns=df_test.columns)
```

# (c) ii.

```python
wl_right = wave_lens[wave_lens >= 1300]
wl_left = wave_lens[wave_lens < 1200]
```

```python
df_fs_tv_r = df_fs_tv[wl_right]
df_fs_tv_l = df_fs_tv[wl_left]
```

```python
df_fs_tv_r.shape
```

```python
def ker(t):
    return np.max(1 - t, 0)
```

```python
# Similar to k-nearest-neighbor algorithm
num_neighb = 3 # number of neighbours to consider
errors = []
preds_tv = []
for k, row in df_fs_tv_r.iterrows():
    dists = ((df_fs_tv_r - row) ** 2).sum(axis=1)
    max_d = dists.max()
    neigh_ds = dists.sort_values()[:num_neighb]

    p1 = np.sum([ker(d / max_d) * df_fs_tv_l.loc[idx] for (idx, d) in neigh_ds.items()], axis=0)
    p2 = np.sum([ker(d / max_d) for (idx, d) in neigh_ds.items()])
    f_left_hat = p1 / p2
    preds_tv.append(f_left_hat)
    error = np.sum((f_left_hat - df_fs_tv_l.loc[k]) ** 2)
    errors.append(error)
```

```python
print(np.mean(errors))
```

Visualize some predictions

```python
fig, axes = plt.subplots(3, 3, figsize=(14, 12))
axes = axes.ravel()

for k, idx in enumerate([0, 5, 10, 15, 20, 25, 30, 35, 40]):
    ax = axes[k]
    ax.plot(wave_lens, df_fs_tv.loc[idx], label='smoothed')
    ax.plot(wl_left, preds_tv[idx], label='predicted')
    ax.legend()
    ax.set_title('Example {0}'.format(idx + 1))
plt.tight_layout()
plt.show()
```
